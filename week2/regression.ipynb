$\hat{\beta} = (X^\prime X)^{-1}X^\prime Y$

```python
X = np.array(df.drop(columns = ['(혈청지오티)ALT']))
Y = np.array(df['(혈청지오티)ALT'])

beta_hat = np.matmul(np.matmul(np.linalg.inv(np.matmul(X.T,X)),X.T),Y)
```

```python
beta_hat
```

$\hat{Y} = X \beta$

```python
predict = np.matmul(X, beta_hat)
```

$\epsilon = Y - \hat{Y}$

```python
error = Y - predict
```

```python
SST = np.sum(np.power(Y-predict.mean(),2))
SSE = np.sum(np.power(Y-predict,2))
SSR = np.sum(np.power(predict-predict.mean(),2))
R_squared = SSR/SST
print(R_squared)
```
